```python
import jinja2
import pandas as pd
from pycaret.classification import setup, compare_models, evaluate_model, predict_model, save_model, load_model
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
s = setup(data, target='Survived')
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
test = pd.read_csv('test.csv')
prediction = predict_model(best_model, data=test)

prediction.head()
```

# Submission File Format:

You should submit a csv file with exactly 418 entries plus a header row. Your submission will show an error if you have extra columns (beyond PassengerId and Survived) or rows.

The file should have exactly 2 columns:

**PassengerId** (sorted in any order) <br>
**Survived** (contains your binary predictions: 1 for survived, 0 for deceased)

```python
answer = prediction.rename(columns={'Label': 'Survived'})
answer = answer[['PassengerId','Survived']]
answer
```

```python
answer.to_csv('Titanic_Submission.csv')
```
